```python
## Regression problem 

### model => close_price(t) = c0 + c1*open_price(t-1) + c2*open_price(t) + c3*close_price(t-1) + c4*high_price(t-1) +  c5*low_price(t-1) + c6*volume(t-1)
```

```python
price_ts['prev_open'] = price_ts['Open'].shift(1)
price_ts[['Open', 'prev_open', 'Close']].head()
```

```python
price_ts['prev_close'] = price_ts['Close'].shift(1)
price_ts['prev_high'] = price_ts['High'].shift(1)
price_ts['prev_low'] = price_ts['Low'].shift(1)
```

```python
price_ts['prev_volume'] = price_ts['Volume'].shift(1)
```

```python
price_ts['prev_close_10'] = price_ts['Close'].shift(5)
price_ts.loc["2023-01-01":, ['Close', 'prev_close_10']].plot()
plt.show()
```

```python
price_ts.head()
```

```python
price_ts = price_ts.drop(["prev_close_10"], axis=1)
price_ts.head()
```

```python
target = 'Close'
input_variables = ['Open', "prev_open", "prev_close", "prev_high", "prev_volume"]
```

```python
price_ts['constant']=1
```

```python
train_df = price_ts.loc[price_ts.index<'2023-01-01']
train_df = train_df.dropna()
val_df = price_ts.loc[price_ts.index>='2023-01-01']

train_df.shape, val_df.shape
```

```python
reg = sm.OLS(train_df[target], train_df[input_variables+['constant']])
model = reg.fit()
```

```python
model.summary()
```

```python
pred_train = model.predict(train_df[input_variables+['constant']])

plt.plot(pred_train)
plt.plot(train_df[target])
plt.show()
```

```python
preds = model.get_prediction()
preds.summary_frame().loc["2022-12-01":, ['mean', "mean_ci_lower", "mean_ci_upper"]].plot()
plt.show()
```

```python
pred_val = model.predict(val_df[input_variables+['constant']])

plt.plot(pred_val)
plt.plot(val_df[target])
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(val_df[target], pred_val)
```

```python
mean_absolute_error(train_df[target], pred_train)
```

```python
### arima model 
### close_price(t) = close_price(t-1), close_price(t-2), .. ma(close_price(t-1)), ma(close_price(t-2)),..
```

```python
aic_full = {}
for p in range(1, 5):
    for q in range(1, 5):
        model = sm.tsa.statespace.SARIMAX(train_df['Close'], order=(p,1,q), enforce_invertibility=False)
        res = model.fit(disp=False)
        aic_full[f"{p}_{q}"] = res.aic
```

```python
import pandas as pd
from tqdm import tqdm

pred = []
true_y = []

val_df = val_df.loc[:'2023-01-31']

for i, row in tqdm(val_df.iterrows(), total = val_df.shape[0]):
    model = sm.tsa.statespace.SARIMAX(train_df['Close'], order=(4,0,2), enforce_invertibility=False)
    res = model.fit(disp=True)
    pred.append(res.forecast(1))
    true_y.append(row['Close'])
    train_df = pd.concat([train_df, row], axis=0)
```

```python
plt.plot(pred)
plt.plot(true_y)
plt.show()
```
