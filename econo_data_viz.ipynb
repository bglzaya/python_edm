```python
for col in df_quarterly.columns[1:]:
    print("*"*10, col, "*"*10)
    print('Нийт мөрийн тоо:', df_quarterly.shape[0])
    print('Нийт хоосон утгын тоо:',  df_quarterly[col].isnull().sum())
    print('Өгөгдлийн төрөл:',  df_quarterly[col].dtype)
    print('Хамгийн их утга:',  df_quarterly[col].max())
    print('Хамгийн бага утга:',  df_quarterly[col].min())
    print('Дундаж утга:',  df_quarterly[col].mean())
    print('Голч утга:',  df_quarterly[col].median())
```

```python
#GDP BY AIMAG

df_gdp = pd.read_excel("data_econo/GDP_AIMAG.xlsx", header=1)
df_gdp
```

```python
import geopandas as gpd

geodf = gpd.read_file("data_grazing/Adminstrative_shape_file/Aimag_name_1.shp")
geodf = geodf.rename(columns={"TextString":"Aimag"})
print(geodf.columns)
```

```python
geodf['Aimag'].sort_values(), df_gdp["Аймаг"].sort_values()
```

```python
geodf = geodf.to_crs("WGS84").set_index("Aimag")
```

```python
year_cols = df_gdp.columns[2:].tolist()
year_cols
```

```python
## gdp data convert
df_gdp_new = []
for col in year_cols:
    temp = df_gdp[['Аймаг', col]]
    temp['year'] = col
    temp.columns = ['Aimag', 'GDP', 'YEAR']
    df_gdp_new.append(temp)

df_gdp_new = pd.concat(df_gdp_new, axis=0)
df_gdp_new
```

```python
import plotly.express as px

fig = px.choropleth_mapbox(df_gdp_new,
                           hover_data = "GDP",
                           geojson=geodf.geometry,
                           locations='Aimag',
                           center={"lat": 47.20900020640241, "lon": 102.8187267977119},
                           color='GDP', mapbox_style="open-street-map",
                           color_continuous_scale="brwnyl", color_discrete_sequence=["fuchsia"],
                           zoom=5.2, animation_frame="YEAR")

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0}, transition = {'duration': 1000})
fig.show()
```
