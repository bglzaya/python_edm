```python
import pandas as pd
```

```python
data = pd.ExcelFile("data_econo/Data_all.xlsx")
```

```python
df = []
for sheet in data.sheet_names:
    df.append(pd.read_excel(data, sheet_name= sheet))
```

```python
for dt in df:
    print(dt.head())
```

```python
df_quarterly = df[1]
```

```python
df_quarterly
```

```python
for col in df_quarterly.columns[1:]:
    print("*"*10, col, "*"*10)
    print('Нийт мөрийн тоо:', df_quarterly.shape[0])
    print('Нийт хоосон утгын тоо:',  df_quarterly[col].isnull().sum())
    print('Өгөгдлийн төрөл:',  df_quarterly[col].dtype)
    print('Хамгийн их утга:',  df_quarterly[col].max())
    print('Хамгийн бага утга:',  df_quarterly[col].min())
    print('Дундаж утга:',  df_quarterly[col].mean())
    print('Голч утга:',  df_quarterly[col].median())
```

```python
df_quarterly = df_quarterly.rename(columns = {"Unnamed: 0": "Quarter"})
df_quarterly.set_index(["Quarter"], inplace=True)
```

```python
df_quarterly.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(df_quarterly["gdp_prod"].iloc[-16:], color='green', linestyle="--")
plt.ylabel("GDP")
plt.xlabel("Quarter")
plt.show()
```

```python
fig, ax1 = plt.subplots()

color = 'tab:green'
ax1.set_xlabel('Quarter')
ax1.set_ylabel('ДНБ')
ax1.plot(df_quarterly["gdp_prod"].iloc[-16:], color=color, label='ДНБ')
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')
ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Нүүрс')
ax2.plot(df_quarterly["coal"].iloc[-16:], label='Нүүрс')
ax2.tick_params(axis='y')

fig.tight_layout()
plt.legend()
plt.show()
```
